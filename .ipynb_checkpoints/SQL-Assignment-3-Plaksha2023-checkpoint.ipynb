```python
import sqlite3
import pandas as pd
# Connect to the database (creates the database file if it does not exist)
conn = sqlite3.connect('students.db')

# Create a cursor object
cursor = conn.cursor()

# Create the students table
cursor.execute('''
CREATE TABLE students (
student_id INT,
name TEXT,
major TEXT,
gpa REAL,
enrollment_date TEXT
);
''')

# Insert data into the table
cursor.execute('''
INSERT INTO students (student_id, name, major, gpa, enrollment_date)
VALUES
(1, 'John', 'Computer Science', 3.5, '2022-01-01'),
(2, 'Jane', 'Physics', 3.8, '2022-01-02'),
(3, 'Bob', 'Engineering', 3.0, '2022-01-03'),
(4, 'Samantha', 'Physics', 3.9, '2022-01-04'),
(5, 'James', 'Engineering', 3.7, '2022-01-05'),
(6, 'Emily', 'Computer Science', 3.6, '2022-01-06'),
(7, 'Michael', 'Computer Science', 3.2, '2022-01-07'),
(8, 'Jessica', 'Engineering', 3.8, '2022-01-08'),
(9, 'Jacob', 'Physics', 3.4, '2022-01-09'),
(10, 'Ashley', 'Physics', 3.9, '2022-01-10');
''')

# Commit changes and close the connection
conn.commit()
conn.close()

# Connect to the database
conn = sqlite3.connect('students.db')

# Create a cursor object
cursor = conn.cursor()

# Select all records in the table
select_all_query = 'SELECT * FROM students;'
select_all = pd.read_sql_query(select_all_query, conn)
print('Select all records in the table',select_all)

# Select students whose major is "Computer Science"
select_cs_query = 'SELECT * FROM students WHERE major="Computer Science";'
select_cs = pd.read_sql_query(select_cs_query, conn)
print('Select students whose major is "Computer Science"',select_cs)

# Select all unique majors and order them by name, descending order
select_majors_query = 'SELECT DISTINCT major FROM students ORDER BY major DESC;'
select_majors = pd.read_sql_query(select_majors_query, conn)
print('Select all unique majors and order them by name, descending order',select_majors)

# Select all students that have an 'e' in their name and order them by gpa in ascending order
select_e_query = "SELECT * FROM students WHERE name LIKE '%e%' ORDER BY gpa ASC;"
select_e = pd.read_sql_query(select_e_query, conn)
print('Select all students that have an "e" in their name and order them by gpa in ascending order',select_e)

# Close the connection
conn.close()
```

```python
# First, import the necessary libraries
import sqlite3
import pandas as pd

# Connect to the database and create a cursor
conn = sqlite3.connect('student_db.sqlite')
cursor = conn.cursor()

# Select all records in the table
select_all_query = 'SELECT * FROM students;'
select_all = pd.read_sql_query(select_all_query, conn)
print(select_all)

# Select students whose major is "Computer Science"
select_cs_query = 'SELECT * FROM students WHERE major="Computer Science";'
select_cs = pd.read_sql_query(select_cs_query, conn)
print(select_cs)

# Select all unique majors and order them by name, descending order
select_majors_query = 'SELECT DISTINCT major FROM students ORDER BY major DESC;'
select_majors = pd.read_sql_query(select_majors_query, conn)
print(select_majors)

# Select all students that have an 'e' in their name and order them by gpa in ascending order
select_e_query = "SELECT * FROM students WHERE name LIKE '%e%' ORDER BY gpa ASC;"
select_e = pd.read_sql_query(select_e_query, conn)
print(select_e)

# Close the connection
conn.close()
```
